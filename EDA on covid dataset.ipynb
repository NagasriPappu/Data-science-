## EDA

```python
import seaborn as sns                     # Seaborn is a library for making statistical graphics in Python
import matplotlib.pyplot as plt           # Plotting library for Python
%matplotlib inline
```

```python
total_cases = data[['State_or_UTs','Total Cases']].sort_values(by=['Total Cases'],ascending=False).reset_index(drop=True)
plt.figure(figsize=(10,10))
ax =sns.barplot(x='Total Cases',y='State_or_UTs',data=total_cases)
ax.bar_label(ax.containers[0],padding=10,fmt='%g');
plt.show()
```

##### It can be clearly observed that Maharastra has the highest no.of total cases

```python
#states having the highest number of deaths and lowest number of deaths?
deaths = data[['State_or_UTs','Deaths']].sort_values(by=['Deaths'],ascending=False).reset_index(drop=True)
plt.figure(figsize=(8,10))
ax =sns.barplot(x='Deaths',y='State_or_UTs',data=deaths)
ax.bar_label(ax.containers[0],padding=10,fmt='%g');
plt.show()
```

### State wise Active, Discharged , Total, Deaths cases

```python
for su in data['State_or_UTs'].values:
    df = data[['Total Cases','Active','Discharged','Deaths']][data['State_or_UTs'] == su]
    df.plot(kind='bar',log=True,title=su)
    plt.show()
```
